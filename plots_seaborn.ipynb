```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.__version__)
# https://www.youtube.com/watch?v=TLdXM0A7SR8
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y= "month", data = a)
plt.show()
```

```python
# data from the following github link can be used in seaborn directly
# https://github.com/mwaskom/seaborn-data

a = sns.load_dataset("flights")
sns.relplot(x="passengers", y= "month", hue= "year", data = a)
plt.show()
```

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y= "tip",data = b, kind = "line")
plt.show()
```

```python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=b);
plt.show()
```

```python
sns.catplot(x='day', y = 'total_bill', data = b)
plt.show()
```

```python
sns.distplot(b['total_bill'], kde = False, color ='red', bins = 30) 
plt.show()
```

```python
sns.distplot(b['total_bill'], hist=False, rug=True)
plt.show()
```

```python
bill_data = b['total_bill']
print(len(bill_data), max(bill_data), min(bill_data))
```

```python
sns.distplot(b['total_bill'])
plt.show()
```

```python
# https://seaborn.pydata.org/tutorial/relational.html#relating-variables-with-scatter-plots
sns.relplot(x="total_bill", y="tip", hue="smoker",
            col="time", data=b);
plt.show()
```

```python
sns.distplot(b['total_bill'])
plt.show()
```

```python
sns.set(style="darkgrid")
g = sns.jointplot("total_bill", "tip", data=b,
                  kind="reg", truncate=False,
                  xlim=(0, 60), ylim=(0, 12),
                  color="g", height=7)
plt.show()
```

```python
ordered_days = b.day.value_counts().index
g = sns.FacetGrid(b, row="day", row_order=ordered_days,
                  height=1.7, aspect=4,)
g.map(sns.distplot, "total_bill", hist=False, rug=True);
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=b);
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", data=b);
plt.show()
```

```python
with sns.axes_style("white"):
    g = sns.FacetGrid(b, row="sex", col="smoker", margin_titles=True, height=2.5)
g.map(plt.scatter, "total_bill", "tip", color="#334488", edgecolor="white", lw=.5);
g.set_axis_labels("Total bill (US Dollars)", "Tip");
g.set(xticks=[10, 30, 50], yticks=[2, 6, 10]);
g.fig.subplots_adjust(wspace=.02, hspace=.02);
plt.show()
```
