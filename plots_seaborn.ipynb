```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
print(sns.__version__)
# https://www.youtube.com/watch?v=TLdXM0A7SR8
```

```python
a = sns.load_dataset("flights")
sns.relplot(x="passengers", y= "month", data = a)
plt.show()
```

```python
# data from the following github link can be used in seaborn directly
# https://github.com/mwaskom/seaborn-data

a = sns.load_dataset("flights")
sns.relplot(x="passengers", y= "month", hue= "year", data = a)
plt.show()
```

```python
b = sns.load_dataset("tips")
sns.relplot(x="time", y= "tip",data = b, kind = "line")
plt.show()
```

```python
sns.catplot(x='day', y = 'total_bill', data = b)
plt.show()
```

```python
sns.distplot(b['total_bill'], kde = False, color ='red', bins = 30) 
plt.show()
```
